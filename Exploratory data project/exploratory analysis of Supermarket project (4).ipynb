```python
df_state=df.groupby(['State'])[['Sales','Discount','Profit']].mean()
df_state.head()
```

```python
df_state1=df_state.sort_values('Discount')
df_state1[['Discount']].plot(kind='bar',figsize=(15,5))
plt.title('State wise discount',fontsize=20)
```

```python
df_city=df.groupby(['City'])['Profit','Sales','Discount'].mean()
df_city
```

```python
df_city=df_city.sort_values('Profit')
df_city.head(50)
```

```python
df_city['Profit'].head(50).plot(kind='bar',figsize=(15,5))
plt.title('city wise low profit',fontsize=20)
```

```python
df_city['Profit'].tail(50).plot(kind='bar',figsize=(15,5))
plt.title('gain highest profit in term of city',fontsize=15)
```

```python
df_region=df.groupby(['Region'])[['Profit','Sales','Discount']].mean()
df_region
```
